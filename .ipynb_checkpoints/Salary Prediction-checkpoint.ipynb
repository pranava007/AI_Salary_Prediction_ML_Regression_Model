```python
# pandas is python libray use for data manupulation like table
import pandas as pd
```

```python
# read_csv in one of the pandas method use for read file
# value assign dataset variable
dataset = pd.read_csv("Salary_Data.csv")
dataset
```

```python
# splite the input and output 
# value assign input YearsExperience
independent = dataset[["YearsExperience"]]
independent
```

```python
# splite the input and output 
# value assign input Salary
dependent = dataset[["Salary"]]
dependent
```

```python
# sklearn is a python libray used for ML
# split data for train and test datas 70% and 30% test_size=0.30
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train,x_test,y_train,y_test
```

```python
# modle creation learning process
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
#fit() is a method used caluclation ex: y=wx+b
# learn
regressor.fit(x_train,y_train)
```

```python
weight = regressor.coef_
weight
```

```python
bais = regressor.intercept_
bais
```

```python
# pred test data
y_pred = regressor.predict(x_test)
```

```python
# eveluvation pred and actual data
# nearby 1 good learning modle
from sklearn.metrics import r2_score
r_score = r2_score(y_pred,y_test)
r_score
```

```python
# create file
import pickle 
filename = "finalized_model_linear.sav"
```

```python
# data dump for new file
pickle.dump(regressor,open(filename ,'wb'))
```

```python
# model save 
load_model = pickle.load(open("finalized_model_linear.sav",'rb'))
result = load_model.predict([[20]])
result
```
